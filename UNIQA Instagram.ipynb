## UNIQA Instagram

Instagram shut down its public API in 2018. You now have to be approved through Facebook before hand, and many of the functionality only works if your account has been verified as a business account. Similarly to how I handled Twitter, I will be attempting to use Apify to scrape Instagram. They have an `Instagram Scraper` and an `Instagram Profile Scraper`. *Otherwise, I could go on Instagram and manually grab the posts, comments, and threads myself possibly.*


```python
#importing libraries
import pandas as pd
```

```python
#This was queried based on UNIQA's instagram URL (@uniqa.at)
instagram_uniqa = pd.read_csv('dataset_instagram_scraper.csv')
instagram_uniqa
```

```python
filtered_comments = [comment for comment in instagram_uniqa['firstComment'] if pd.notnull(comment)]
filtered_comments
```

```python
len(filtered_comments)
#269 comments on 482 UNIQA posts
```

So far this has worked surprisingly well actually, and I was able to get all of this data for free. The `Instagram Scraper` on Apify allows up to 5 dollars of free usage and this cost me about $0.60.
